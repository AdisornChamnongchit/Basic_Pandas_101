```python
import pandas as pd
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,12,5,8,3]
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

for i in range(len(x)):
    plt.text(x[i], y[i], str(y[i]), ha='center', va='bottom')

plt.show()
```

# 2. Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
value = [25,30,40,50]
plt.bar(categories,value)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Value")

for i in range(len(categories)):
    plt.text(categories[i], value[i], str(value[i]), ha='center', va='bottom')

plt.show()
```

# 3. Pie Chart

# 4. Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = np.random.randn(500)
plt.hist(data, bins=30,edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

# 6. Box Plot

```python
import seaborn as sns
import numpy as np

data = [np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")

plt.show()
```

# 7. Violin Plot

```python
import seaborn as sns
import numpy as np

data = [np.random.normal(0,std,100)for std in range(1,4)]
plt.violinplot(data)
plt.title("Violin Plot")
plt.xlabel("Box")
plt.ylabel("Value")

plt.show()
```

# 9. Area Chart

# 10. Radar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C', 'D','E']
data = np.array([4,5,3,4,2])
angles = np.linspace(0,2*np.pi,len(labels),endpoint=False)
data = np.concatenate((data,[data[0]]))
angles = np.concatenate((angles,[angles[0]]))

plt.polar(angles,data,'o-')
plt.fill(angles,data,alpha=0.25)
plt.title("Radar Chart")
plt.xticks(angles[:-1],labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ["Python","Scala","C#",'Java','PHP']
index = (1,2,3,4,5)
sizes = [45,10,15,30,22]

plt.bar(index,sizes,tick_label = labels , color = ("red","green","blue","yellow","orange"))

plt.title("Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Popularity")

# Add labels to the bars
for i in range(len(index)):
    plt.text(index[i], sizes[i], str(sizes[i]), ha='center', va='bottom')

plt.show()
```
