```python
import pandas as pd
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,12,5,8,3]
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

for i in range(len(x)):
    plt.text(x[i], y[i], str(y[i]), ha='center', va='bottom')

plt.show()
```

# 2. Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
value = [25,30,40,50]
plt.bar(categories,value)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Value")

for i in range(len(categories)):
    plt.text(categories[i], value[i], str(value[i]), ha='center', va='bottom')

plt.show()
```

# 3. Pie Chart

```python
import matplotlib.pyplot as plt
import pandas as pd

status = ["Pass","Not Pass"]
values = [1289,545]
plt.pie(values, labels=status, autopct="%1.1f%%")
plt.title("Pie Chart")

plt.show()
```

# 4. Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = np.random.randn(500)
plt.hist(data, bins=30,edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

# 5. Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd

x = np.random.rand(50)
y = 2*x+1+0.1*np.random.randn(50)
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()

```

# 6. Box Plot

```python
import seaborn as sns
import numpy as np

data = [np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")

plt.show()
```

# 7. Violin Plot

```python
import seaborn as sns
import numpy as np

data = [np.random.normal(0,std,100)for std in range(1,4)]
plt.violinplot(data)
plt.title("Violin Plot")
plt.xlabel("Box")
plt.ylabel("Value")

plt.show()
```
