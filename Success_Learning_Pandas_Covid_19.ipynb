```python
#อัตราผู้ติดเชื้อ Covid ?
```

```python
pop_df = pd.read_csv('/content/thai_population_63.csv')
pop_df = pop_df[['จังหวัด', 'รวม']]
pop_df
```

```python
province_df = df.groupby('province_of_isolation').count()['No.'].reset_index()
```

```python
province_df
```

```python
df.insert(2, 'province', df['province_of_isolation'].replace('กทม', 'กรุงเทพมหานคร'), True)
province_df = df.groupby('province').count()['No.'].reset_index()
province_df
```

```python
prov_pop_df = pd.merge(province_df, pop_df, left_on='province', right_on='จังหวัด')
prov_pop_df
```

```python
prov_pop_df['rate'] = prov_pop_df['No.'] * 1000000 / prov_pop_df['รวม']
prov_pop_df
```

```python
top10_prov_df = prov_pop_df.sort_values('rate').head(10).set_index('จังหวัด')
top10_prov_df
```

```python
top10_prov_df[['No.', 'rate']].plot.bar()
```
